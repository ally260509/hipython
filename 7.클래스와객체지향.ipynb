# 클래스 - 구조
객체 생성하기 위한 툴
- 객체 구성 요소 : 특성(속성) + 행위(메서드)  * 특성은 변수/행위는 함수

- 기본 메서드를 상속받아서 쓸 수 있다.

```python
class Bicycle:   #python Object 원조부모
  def __init__(self, color, use_case):  #생성자
    self.color = color  #속성
    self.use_case = use_case

  def move(self):  #햠수 > 메서드
    print(f'{self.color} {self.use_case} 자전거가 움직입니다.')
```

```python
b1 = Bicycle('red','산악용')  # 객체생성 
b2 = Bicycle('white','출퇴근용')
```

```python
b1.move()
```

# 자동차 클래스 정의

- self.속성명 : instance 속성
- 속성명 : 클래스 속성

- __init__(self) 생성자, 매개변수 = 인스턴스 속성 설정
- def 메서드(self): 메서드 정의

- 클래스 메서드(cls) : 클래스 속성 관리, @classmethod "데코레이터"

- 정적 메서드 : 유틸리티

```python
class Car:
  #1 - 클래스 속성
  instance_cnt = 0

  def __init__(self, color=''):  #생성자 정의
    #2 - 객체의 속성
    self.color = color

    Car.instance_cnt += 1

    print(f'{self.color} 자동차 객체가 생성되었습니다({Car.instance_cnt}대)')

  def auto_pilot(self):
    print(f'{self.color} 자동차가 자율주행중입니다.')

    print(f'{Car.instance_cnt}대의 자동차가 운행중입니다.')
      #00대 자동차가 운행중입니다.
  @classmethod
  def count_check(cls):
    print(f'자동차 대수: {cls.instance_cnt}')

  @staticmethod
  def type_check(type_code):  #유틸리티
    if type_code > 20:
      print('전기자동차')
    elif type_code < 20:
      print('내연자동차')
    else:
      print('하이브리드')
```

```python
#Q1. 

class Calculator:
  #생성자
  def __init__ (self, number_list):
    #인스턴스 속성 정의
    self.number_list = number_list
  
  #인스턴스 메소드  
  def sum(self):
    total = sum(self.number_list)
    print(f'합계 : {total}')
    return total

  #인스턴스 메소드
  def avg(self):
    avg_value = self.sum()/len(self.number_list)
    print(f'평균 : {avg_value:.2f}')
    
```

```python
c1 = Calculator([1,2,3,4,5])
c1.number_list
```

```python
c2 = Calculator([6,7,8,9,10])
c2.number_list
```

```python
c1.sum()
c2.sum()
```

```python
c1.avg()
c2.avg()
```

#Q2.Account 클래스 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Acoount 클래스를 생성한 후 생성자를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력받습니다. 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. 

#은행이름: SC은행
#계좌번호: 111-11-111111

#클래스: Account
#생성자: 예금주(owner), 초기 잔액(balance)
#자동설정: 은행이름(sc은행), 계좌번호(모두 랜덤 숫자)
#속성: bank_name, owner, account_number, balance

```python
import random

class Account:
  def __init__(self, owner, balance):
    self.owner = owner
    self.bank_name = 'SC은행'
    self.balance = balance
    self.account_number = Account.makeAccount()
  # staticmethod 호출 
  @staticmethod
  def makeAccount():
    num1 = str(random.randint(0,999)).zfill(3)
    num2 = str(random.randint(0,99)).zfill(2)
    num3 = str(random.randint(0,999999)).zfill(6)
    return f'{num1}-{num2}-{num3}'
  
  #인스턴스 메소드
  def displayAccount(self):
    print(f'은행명 : {self.bank_name}')
    print(f'예금주 : {self.owner}')
    print(f'계좌번호 : {self.account_number}')
    print(f'잔액 : {self.balance}')
  
  #입금 기능
  def deposit(self, amount):
    if amount <= 0:
      print('입금금액은 0보다 커야 합니다.')
      return
    
    self.balance += amount
    print(f'{amount}원 입금완료. 현재 잔액: {self.balance}원')
  
  #출금 기능
  def withdraw(self, amount):
    if amount <=0:
      print('출금금액은 0보다 커야 합니다.')
      return
    
    self.balance -= amount
    print(f'{amount}원 출금 완료. 현재 잔액: {self.balance}원')
```

```python
acc1 = Account('최아영', 1000)
acc1.displayAccount()
acc1.deposit(50000)
acc1.withdraw(30000)
```
