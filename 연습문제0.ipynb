```python
#3. 유저한테 두개의 float을 입력받아서, 두 수의 곱과 합을, 다음처럼 똑같이 출력하는 코드를 만드세요.
    # 출력형태: First is 4.3, second is 5.21. So, Mutiplication = 22.403, addition = 9.51

float_one = float(input("소수점 한자리 숫자를 입력해주세요: "))
float_two = float(input("소수점 두자리 숫자를 입력해주세요: "))
multi = float_one * float_two
add = float_one + float_two
print(f"First is {float_one}, second is {float_two}. So, Mutiplication = {multi}, addition = {add}")
```

```python
num_1 = float(input('실수(예: 1.1)를 입력하세요'))
num_2 = float(input('실수(예: 1.222)를 입력하세요'))

mul = num_1 * num_2
addition = num_1 + num_2
output = f'First is {num_1}, second is {num_2}.'\
        + f'So, Mutiplication = {mul:.3f},'\
        + f'addition = {addition:.2f}'
print(output)
```

```python
#6. 어떤 숫자 x가 주어졌을 때, 그 숫자가 짝수인지 홀수인지 계산하는 코드를 작성하세요.

x = int(input("숫자를 입력해주세요: "))
if x // 2 == 0:
    print("짝수입니다")
else:
    print("홀수입니다")
```
