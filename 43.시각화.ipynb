```python
import folium

# 1) 서울 중심(시청 근처)으로 지도 생성
seoul_center = [37.5665, 126.9780]
m = folium.Map(location=seoul_center, zoom_start=12, tiles="OpenStreetMap")

# 2) 표시할 명소 5곳 (이름, 위도, 경도)
spots = [
    {"name": "경복궁", "lat": 37.579617, "lon": 126.977041},
    {"name": "N서울타워(남산타워)", "lat": 37.551169, "lon": 126.988227},
    {"name": "명동", "lat": 37.563757, "lon": 126.985302},
    {"name": "동대문디자인플라자(DDP)", "lat": 37.566295, "lon": 127.009356},
    {"name": "롯데월드타워(잠실)", "lat": 37.513068, "lon": 127.102502},
]

# 3) 마커 추가
for s in spots:
    folium.Marker(
        location=[s["lat"], s["lon"]],
        popup=s["name"],
        tooltip=s["name"],
        icon=folium.Icon(color="blue", icon="info-sign"),
    ).add_to(m)

# 4) 저장 (주피터에서는 마지막 줄에 m만 적어도 렌더링됨)
m.save("seoul_spots_map.html")
m
```
