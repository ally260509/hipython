````python
# 프롬프트 호출 > 응답# 프롬프트 호출 > 응답
res = client.responses.create(
    model = 'gpt-4o-mini',
    input = [
        {'role':'system','content':'너는 마케팅 전문가야'}, 
        {'role':'user','content':'광고 문구를 작성'}
    ],
    temperature = 0
)
````

# 헬퍼 함수 - input_text 유형의 메시지 처리

````python
# 헬퍼 (코드의 안정성을 위해 추가한 함수)
def msg(role, text) :
    msg_src = {'role': role, 'content': [{'type':'input_text', 'text': text}]}
    return msg_src
````

# 이전 대화를 포함하여 답변하기

````python
response = client.responses.create(
    model = 'gpt-4o-mini',
    input = [
        msg('user', '2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?'),
        msg('user', '그 나라가 화제가 되었던 이유는?')
    ],
    temperature = 0
)

response.output_text
````

# 헬퍼함수 - output_text 유형의 메시지 처리
- system
- user
- assistant : 그 전에 기억하고 싶은 대화를 저장 -> 래그로, 사람이 검수, 중간산출물 넣을 때 사용

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        {
            "role": "assistant", #헬퍼는 그대로 두고, assistant만 명시적으로 처리
            "content": [
                {
                    "type": "output_text",
                    "text": "바로 예상을 뚫고 우승한 브라질입니다."
                }
            ]
        },
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
````

````python
response.output_text
````

# 파라미터의 제어
- temperature : 창의성 0~1 (최대)
- max_output_tokens : 최대 응답 길이
- top_p :    # 정확도이므로 temperature와 top_p는 상충되므로 동시에 지정해주지 않음

````python
top_p = [0.6, 0.8, 1.0]
for p in top_p :
    response = client.responses.create(
        model = 'gpt-4o-mini',
        input = '마케팅 문구 1개 만들어줘',
        top_p = p
    )
    print(response.output_text)
````

````python
temperature = [0.1, 0.5, 1.0]
for t in temperature :
    response = client.responses.create(
        model = 'gpt-4o-mini',
        input = '마케팅 문구 1개 만들어줘',
        temperature = t
    )
    print(response.output_text)
````
