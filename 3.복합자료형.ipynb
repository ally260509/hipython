# 리스트와 for문

```python
students = [{'num':'1','name':'kim','kor':80,'math':90,'eng':100,},
            {'num':'2','name':'choi','kor':85,'math':80,'eng':95},
            {'num':'3','name':'park','kor':100,'math':60,'eng':90}]
students[2]['name']  #인덱싱하기
```

```python
print(students)
```

```python
print(students[0])  #for문으로 간단히!!
print(students[1])
print(students[2])
```

```python
for s in students:
    s['total'] = s['kor'] + s['math'] + s['eng']   
    print(s['total'])
```

```python
print(students)    #total 변수가 들어감
```

```python
for s in students:
    s['average'] = s['total'] / 3
    print(f'{s['average']:.1f}')   # 출력화면만 소수점 1자리로 보여짐
```

```python
for s in students:
    s['average'] = round(s['total']/3,2) # 데이터 자체 소수점 자리를 바꿔줄 때
    print(s)                              
```

```python
for s in students:
    print(s)
```
