```python
def calculate_sum(a, b):
    result = a + b # 들여쓰기된 코드 블록
    return result  # 같은 레벨의 들여쓰기

result = calculate_sum(3, 4)

if result > 0:
    print(result)
    print("양수입니다")  # if문 내부의 들여쓰기 된 코드
```
