# 인덱싱

[] 인덱싱

- iloc[] 행기반 인덱싱 >>> [행,열]
- loc[] 이름기반(위치) 인덱싱 >>> [행이름,열이름]

- : >> start:end까지 슬라이싱
- loc[행이름:행이름]
- iloc[행번호:행번호] >>> end 행번호는 -1 까지

```python
#df1[['이름', 'age']]
df1.loc[:,['이름', 'age']]  #loc는 행과 열을 모두 지정해주어야 함 >> loc[행, 열]
```

```python
df1.loc['A']  #행만 꺼내면 Series로 나옴
```

```python
df1.iloc[0:3, 0:2] #행 번호, 열 번호 (,를 기준으로 행과 열을 지정함) (:은 start:end(-1))
```
