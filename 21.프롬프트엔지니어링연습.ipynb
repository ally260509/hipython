```python
#프롬프트 실행함수
def return_label(text) :  # text : 고장내용
    response = client.responses.create(
        model = 'gpt-4o-mini',
        input = [
            msg('system', sys_prompt),
            msg('user', '고장내용:' + text + '\n출력:')
        ],
        temperature = 0
    )
    return response.output_text  #출력 리턴
```

```python
#프롬프트 호출해서 결과를 리스트에 추가
result = []
for item in sub_df['고장내용']:
    res_output = return_label(item)  #프롬프트 호출 - 1회용 함수 활용
    result.append(res_output) #결과를 리스트에 추가
```
