# CRUD : 간식박스 프로그램

1. 함수 도출 : CRUD 기본

```python
#간식 박스 생성 함수
def make_snackbox():
    snackbox = ['두쫀쿠', '홈런볼']
    return snackbox
```

```python
#간식 박스 추가 함수
def add_snackbox(sbox):  #간식리스트
    make_msg = input('먹고 싶은 간식을 추가하세요, 쉼표(,)로 연결하세요').split(',')
    sbox.extend(make_msg)
    #print(sbox)
    return sbox
```

```python
def display_snackbox(sbox):
    print(sbox)
```

```python
#메뉴출력 함수
def display_menu():
    print('1. 간식추가')
    print('2. 간식박스 확인')
    print('3. 간식 제거')
    print('4. 주문확정')
```

```python
def order_confirm():
    make_msg = input('주문을 확정하시겠습니까? (y/n)')
    if make_msg == 'y' :
        return True
    else :
        return False
```

```python
#전체 보기 함수
#삭제 함수
#수정 함수
```

```python
#### 서비스의 흐름 로직
snack_box = make_snackbox()
print('간식박스 생성완료', snack_box)
while True :
    display_menu() #
    menu_no = input('번호를 선택하세요')
    if menu_no == '1' :
        add_snackbox(snack_box)  #간식리스트
    elif menu_no == '2' :
        display_snackbox(snack_box)
    elif menu_no == '4' :
        display_snackbox(snack_box)
        result = order_confirm()  #매개변수 통일
        if result : 
            break 
print('고객님의 주문을 감사드립니다')
```

```python
#함수정의
#피보나치 수열 특징 : 첫 두 숫자는 0과 1
#계산식: c = a + b
def fibo(n):
    fib = []

    for i in range(n):
        if i == 0:
            fib.append(0)
        elif i == 1:
            fib.append(1)
        else:
            fib.append(fib[i-1] + fib[i-2])

    return fib
```
