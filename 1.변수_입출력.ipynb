# 변수의 영역 nameSpace

지역변수, 전역변수

```python
def local_exam():    #사용자정의함수
    local_var = 100
    print(f"함수내부: {local_var}")
local_exam()        #함수 호출
local_var = 200
local_exam()        #5번줄에서 local_var=200을 입력했어도, local_exam()은 def 함수에서 정의한 100을 불러옴
print(f"함수외부: {local_var}")    #local_var=200을 불러옴(전역변수)
```

```python
counter = 0
def inc_counter():
    global counter
    counter += 1     #counter = counter + 1과 동일한 표현 '+='
    print(counter)
    
inc_counter()
inc_counter()
inc_counter()
```
