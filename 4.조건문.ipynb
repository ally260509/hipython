# 학점 판별기
90 : A
80 : B
70 : C
60 : D
나머지: F

```python
#중첩 if문 (elif)
score = int(input('학점을 입력해주세요:'))
if score >=90 :
    print('A입니다')
elif score >=80 :
    print('B입니다')
elif score >=70 :
    print('C입니다')
elif score >= 60 :
    print('D입니다')
else :
    print('F입니다')    
```

```python
#중첩 if문
score = int(input('학점을 입력해주세요:'))
if score >=90 :
    print('A입니다')
else:
    if score >=80 :
        print('B입니다')
    else:
        if score >=70 :
            print('C입니다')
        else:
            if score >= 60 :
                print('D입니다')
            else :
                print('F입니다')    
```

```python
# 계산기 만들기
# 숫자 + 숫자
# 결과입니다. 출력
# +, -, *, /
```

```python
num1, opr, num2 = input('수식을 입력하세요(예. 20 + 20):').split()

num1 = float(num1)
num2 = float(num2)

result = 0
if opr == '+' :
    result = num1 + num2
elif opr == '-' :
    result = num1 - num2
elif opr == '*' :
    result = num1 * num2
elif opr == '/' :
    result = num1 / num2
else:
    print('사칙 연산을 지원합니다')
print(f'{num1} {opr} {num2}의 결과는 {result}입니다.')
```

```python
#재고관리
inventory = {
    '노트북':5, 
    '마우스':120,
    '키보드':0,
    '모니터':15
}
# 0 품절, 10< 부족 <50 적정, >50 충분
for name, qty in inventory.items() :

    if qty <=0 :
        status ='품절'
    elif qty <=10 :
        status = '부족'
    elif qty <=50 :
        status = '적정'
    else :
        status = '충분'
    print(f'{name} : {qty}로 {status} 상태입니다.')
```

```python
sentence = input('문장을 입력하세요: ')
words = ['파이썬', '코딩', '데이터', 'AI']

found_word = None

for word in words:
    if word in sentence:
        found_word = word
        break

if found_word:
    print(f'입력한 문장에서 {found_word} 단어를 찾았습니다')
else:
    print('입력한 문장에 해당 단어가 없습니다')
```

```python
def check_access(user_role) :
    if user_role == 'admin' :
        return '관리자 페이지에 접근가능합니다.'
    elif user_role == 'user' :
        return '일반 사용자 페이지에 접근가능합니다.'
    else:
        return '접근 권한이 없습니다.'

print(check_access('admin'))
print(check_access('user'))
print(check_access('guest'))
```

# break, continue, pass

```python
numbers = [1, 3, 5, 7, 9, 11, 13, 15]
s_value = 9
for n in numbers :
    if (n == s_value) :
        print(f'찾았습니다. {n}')
        break #가장 가까운 반복문
    else :
        print(f'현재 검사한 숫자는: {n}')
```

```python
for i in range(1,11):  #짝수만 출력
    if ((i % 2) == 1): #홀수이면 continue
        continue
    print(i)
```

```python
numbers = [1, 3, 5, 7, 9, 11, 13, 15]
s_value = 9
idx = 1
for n in numbers :
    if(n == s_value):
        print(f' {idx}번째에서 찾았습니다. {n}')
        break
    #else :
    print(f'현재 검사한 숫자는: {n}')
    idx = idx + 1
```
