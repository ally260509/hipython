```python
# 문자열 순회
isUpper = False
isLower = False
isDigit = False
isSpecial = False
special_char = '!@#%?$'

password = 'anC21!'
# 비밀번호 조건 만족여부 확인
for char in password :
    if char.isupper() : #대문자인지 확인
        isUpper = True
    elif char.islower() :
        isLower = True
    elif char.isdigit() : #숫자인지 확인
        isDigit = True
    elif char in special_char :
        isSpecial = True
    check_result = isUpper + isLower + isDigit + isSpecial
print(f"비밀번호 강도 조건 만족여부 : {'만족' if check_result ==4 else '불만족'}")
```

```python
# 연도 날수 계산
total_days = 0
for i in range(1,year):   #2024년까지
    if i % 400 == 0:   #윤년
        total_days += 366
    elif i % 100 == 0 :  #평년
        total_days += 365
    elif i % 4 == 0 :  #윤년
        total_days += 366
    else :  #평년
        total_days += 365
print(f'1st : {total_days}, {year}')
# 2025년의 월기준 일수 계산
month_days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
for m in range(1,month) :
    total_days += month_days[m]

print(f'2nd : {total_days}, {year}')

#올해가 윤년인지 체크 +1
if month >= 3 : 
    #윤년체크
    if year % 400 == 0:   #윤년
        total_days += 1
    elif year % 100 == 0 :  #평년
        total_days += 0
    elif year % 4 == 0 :  #윤년
        total_days += 1
    else :  #평년
        total_days += 0

print(f'3rd : {total_days}, {year}, {month}')

total_days += day

print(f'final : {total_days}, {year}, {month}, {day}')

# 연도 + 월 + 일 % 7 > 요일에 해당하는 수
remain = total_days % 7
result = None
if remain == 0:
    result = '일요일'
elif remain == 1:
    result = '월요일'
elif remain == 2:
    result = '화요일'
elif remain == 3:
    result = '수요일'
elif remain == 4:
    result = '목요일'
elif remain == 5:
    result = '금요일'
elif remain == 6:
    result = '토요일'

print(f'{year}년 {month}월 {day}일은 {result}입니다.')
```

```python
#chatGPT 버전
students = [{'num':'1','name':'kim','kor':80,'math':90,'eng':100,},
            {'num':'2','name':'park','kor':90,'math':60,'eng':80},
            {'num':'3','name':'hong','kor':70,'math':70,'eng':80},
            {'num':'4','name':'hong1','kor':70,'math':70,'eng':80},
            {'num':'5','name':'hong2','kor':70,'math':70,'eng':80}]

# 학생별 총점과 평균 계산(속성 추가)
for s in students :
    total = s['kor'] + s['math'] + s['eng']
    avg = total / 3

    s['total'] = total
    s['avg'] = avg

# 학생별 순위 계산(속성 추가)
for s in students:
    rank = 1
    for other in students:
        if other['total'] > s['total'] :
            rank +=1
    
    s['rank'] = rank

#전체 총점과 평균 계산
all_total = 0

for s in students :
    all_total += s['total']

all_avg = all_total / (len(students) * 3)

#과목별 평균 출력
kor_sum = 0
math_sum = 0
eng_sum = 0

for s in students:
    kor_sum += s['kor']
    math_sum += s['math']
    eng_sum += s['eng']

kor_avg = kor_sum / len(students)
math_avg = math_sum / len(students)
eng_avg = eng_sum / len(students)

print("번호 이름 국어 수학 영어 총점 평균 순위")
for s in students:
    print(s['num'], s['name'], s['kor'], s['math'], s['eng'],
          s['total'], f'{s['avg']:.1f}',s['rank'])

print('\n전체 총점:',all_total)
print('전체 평균:', f'{all_avg:2f}')

print('\n과목별 평균')
print('국어 평균:', f'{kor_avg: 1f}')
print('수학 평균:', f'{math_avg: 1f}')
print('영어 평균:', f'{eng_avg: 1f}')
```

```python
#변수의 초기화
i = 1
sum = 0
while True : #조건
    sum += i  #sum = sum + i
    i += 1   # i = i + 1      #(sum, i) 값이 (45, 10)
    if i >=10 : #조건식
        break
    #i += 1   #(sum, i) 값이 (55,10)
print(sum, i)
```

# 업앤다운 게임

```python
import random
number = random.randint(1,10)
number #컴퓨터의 선택
```

```python
# 정답 맞추기
cnt = 0
while True:
    #입력 받기
    answer = int(input('1~10 사이에 숫자 맞추세요'))
    cnt += 1
    #조건
    if answer == number :  #맞춘 경우
        print(f'축하합니다. 정답입니다. 도전횟수: {cnt}')
        break
    elif answer > number :
        print('down')
    else :
        print('up')    
```
