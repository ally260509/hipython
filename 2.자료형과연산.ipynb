# 문자열

1. 인덱싱, 슬라이싱, 멤버십
2. 문자열은 immutable "불변", 대소문자 구분한다.
3. 문자열은 코드를 지정할 수 있어 인코딩, 디코딩, 유니코드(이모지), 한글 가능하다.
4. 문자열 +, *
5. 공백처리, 형식처리 등을 해야 한다.

```python
import re   #정규표현식을 처리

text = '파이썬 문자열    처리'
#text.split(' ')
re.sub(r'\s+',' ',text)   #\s는 space를 나타내며, +를 붙이는 이유는 공백이 여러 칸일 수 있기 때문임
```

```python
tel_no = '010-123-1225'
re.sub(r'\D','',tel_no)
```
